```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

corruptedData = pd.read_csv('eksikveriler.csv')
print(corruptedData)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
age = corruptedData.iloc[:, 1:4].values

imputer = imputer.fit(age[:, 1:4])
age[:, 1:4] = imputer.transform(age[:, 1:4])
print(age)

country = corruptedData.iloc[:, 0:1].values
print(country)
```

```python
from sklearn import preprocessing
import pandas as pd

# Assuming 'corruptedData' is already loaded with your dataset
# Label encoding the 'country' column
le = preprocessing.LabelEncoder()

country = corruptedData.iloc[:, 0:1].values
country[:, 0] = le.fit_transform(corruptedData.iloc[:, 0])
print(country)

# One-hot encoding the 'country' column
ohe = preprocessing.OneHotEncoder()
country = ohe.fit_transform(country).toarray()
print(country)

# Convert the encoded data into a DataFrame
sonuc = pd.DataFrame(data=country, index=range(len(country)), columns=['fr', 'tr', 'us'])
print(sonuc)

# Create DataFrame for 'age'
age = corruptedData.iloc[:, 1:4].values
sonuc2 = pd.DataFrame(data=age, index=range(len(age)), columns=['boy', 'kilo', 'yas'])
print(sonuc2)

# Encode 'cinsiyet' column directly
cinsiyet = corruptedData.iloc[:, -1].replace({'e': 1, 'k': 0}).values
print(cinsiyet)

# Convert 'cinsiyet' into a DataFrame
sonuc3 = pd.DataFrame(data=cinsiyet, index=range(len(cinsiyet)), columns=['cinsiyet'])
print(sonuc3)

# Concatenate the DataFrames
s = pd.concat([sonuc, sonuc2], axis=1)
print(s)

s = pd.concat([s, sonuc3], axis=1)
print(s)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(s, sonuc3, test_size=0.33, random_state=0)

print(x_train)
print(x_test)
print(y_train)
print(y_test)

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(x_train)
X_test = sc.fit_transform(x_test)
```
